##### A python to get absolute value of a number

```python
t= float(input("write a number"))
x= 17
f=t-x
if t<x:
    print(abs(f))
else:
    print(f*2)
```

```python
a= int(input("write a number"))
b= int(input("write a number"))
c= 3
sum= a+b+c
if a!= 3 and b!=3:
    print(sum)
else:
    print(sum*3)
    
```
